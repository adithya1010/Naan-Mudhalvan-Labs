### Seperating the dataset into features and labels

```python
# Seperate the dataset into features (X) and labels (Y)
X=flight_data_subset_dummies.drop('satisfaction',axis=1)
Y=flight_data_subset_dummies['satisfaction']
```

### Seperating data into training, testing and validation sets

```python
# Separate into train, validate, test sets.
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

### Defining set of hyperparameters

```python
# Determine set of hyperparameters.
   # Update the values for 'min_samples_leaf' and 'min_samples_split' to be within acceptable ranges.
cv_params = {'n_estimators' : [50,100],
                 'max_depth' : [10,50],
                 'min_samples_leaf' : [1, 2],  # Changed to integers or small floats
                 'min_samples_split' : [2, 5],  # Changed to integers or small floats
                 'max_features' : ["sqrt"],
                 'max_samples' : [.5,.9]}
```

validation_size = int(0.2 * len(X_train))

This calculates the size of your validation set. It takes 20% of the number of samples in your training data (X_train) and converts it to an integer. The result, validation_size, represents how many samples will be designated for validation.
Important Note: It's crucial that this calculation is based on len(X_train) after you've already split your data into training and testing sets. If you use len(X) (the original dataset), your validation set will be much larger and overlap with your test set, leading to incorrect evaluation.
split_index = np.full(len(X_train), -1)

This creates a NumPy array named split_index. The array has the same length as your training data (X_train).
np.full(len(X_train), -1) initializes all elements of the array with the value -1. In the context of PredefinedSplit, a value of -1 indicates that a sample belongs to the training set.
split_index[:validation_size] = 0

This modifies the split_index array. It takes the first validation_size elements (i.e., the first 20% of your training data) and sets their values to 0.
In the context of PredefinedSplit, a value of 0 indicates that a sample belongs to the validation set. So, you're essentially marking the first 20% of your training data as the validation set.
custom_split = PredefinedSplit(split_index)

This creates a PredefinedSplit object, which is a special type of cross-validation splitter that tells scikit-learn exactly which samples to use for training and which to use for validation, based on the split_index array.
The custom_split object can then be used as the cv argument in functions like cross_val_score, GridSearchCV, or RandomizedSearchCV to ensure that your model is evaluated using your predefined split.

### Creating list of split indices

```python
# Create list of split indices.
   # Ensure that some data points are assigned to the validation set (index 0).
   # For example, you can randomly assign 20% of the training data to the validation set for the custom split.
import numpy as np

validation_size = int(0.2 * len(X_train))
split_index = np.full(len(X_train), -1)  # Initialize with -1
split_index[:validation_size] = 0  # Assign 0 to the first 20% of data points
custom_split = PredefinedSplit(split_index)
```

**Random Forest Classifier:**
A Random Forest Classifier is an ensemble learning method that combines multiple decision trees to improve the accuracy and robustness of predictions. It works by training multiple trees on random subsets of the data and then aggregating their predictions to produce a final outcome. This approach helps to reduce overfitting, handle high-dimensional data, and improve model interpretability. Random Forest Classifiers are widely used for classification tasks, such as image classification, text classification, and sentiment analysis, and are particularly effective when dealing with complex and non-linear relationships between features and target variables.

**Ensemble Learning:**

Ensemble learning is a machine learning approach that combines the predictions from multiple models to improve the overall accuracy and performance of a model. This method involves training multiple models on the same dataset, each with a different set of parameters or algorithms, and then combining their predictions to produce a single output. Ensemble learning can be used to reduce overfitting, improve robustness, and increase the overall accuracy of a model. Some common ensemble learning methods include bagging, boosting, and stacking.

### Instantiating Random Forest Model

```python
# Instantiate model.

rf = RandomForestClassifier(random_state=0)
```

### Performing GridSearchCV over the specified parameters

```python
# Search over specified parameters.

rf_val = GridSearchCV(rf, cv_params, cv=custom_split, refit='f1', n_jobs = -1, verbose = 1)
```

### Fitting the model

```python
%%time

# Fit the model.

rf_val.fit(X_train, y_train)
```

### Obtaining optimal parameters

```python
# Obtain optimal parameters.

rf_val.best_params_
```

### Using the optimal parameters

```python
# Use optimal parameters on GridSearchCV.

rf_opt = RandomForestClassifier(n_estimators = 50, max_depth = 50,
                                min_samples_leaf = 1, min_samples_split = 0.001,
                                max_features="sqrt", max_samples = 0.9, random_state = 0)
```

A fitting model refers to a mathematical representation or statistical model that accurately describes the relationship between variables in a dataset. In machine learning, a fitting model is one that is well-suited to the specific problem or task at hand, such as predicting continuous outcomes or classifying categorical outcomes. A good fitting model will capture the underlying patterns and relationships in the data, allowing for reliable predictions and decisions to be made. Fitting a model involves adjusting its parameters to minimize the difference between predicted and actual values.

### Fitting the optimal model

```python
# Fit the optimal model.

rf_opt.fit(X_train, y_train)
```

### Predicting on test set

```python
# Predict on test set.

y_pred = rf_opt.predict(X_test)
```

##Evaluating the Model

### Getting recall score

```python
# Get recall score.

rc_test = recall_score(y_test, y_pred, pos_label = "satisfied")
print("The recall score is {rc:.3f}".format(rc = rc_test))
```

### Getting accuracy score

```python
# Get accuracy score.

ac_test = accuracy_score(y_test, y_pred)
print("The accuracy score is {ac:.3f}".format(ac = ac_test))
```

### Getting F1 score

```python
# Get F1 score.

f1_test = f1_score(y_test, y_pred, pos_label = "satisfied")
print("The F1 score is {f1:.3f}".format(f1 = f1_test))
```
